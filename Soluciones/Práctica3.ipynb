## Ejercicio 1

Sea $X$ una v.a. con función de densidad:

$
f_X(x) =
\begin{cases}
  0.75(1-x^2) & -1 \leq x \leq 1 \\
  0 & \text{en otro caso}
\end{cases}
$

### Pregunta A

Verificar que $f_X$ es realmente una función de densidad.

Una función es una función de densidad (PDF) si cumple con 2 cosas:

1. $f_X(x) \geq 0$ para todo $-1 \leq x \leq 1$
2. $\int_{-1}^{1} f_X(x) dx = 1$

$f_X(x) = 0.75(1-x^2)$

$0.75(1-x^2) \geq 0 \iff 1-x^2 \geq 0 \iff 1 \geq x^2 \iff 1 \geq \pm x \iff -1 \leq x \leq 1$

$\int_{-1}^{1} f_X(x) dx = \int_{-1}^{1} 0.75(1-x^2) dx = 1$

### Pregunta B

Calcular: 

$P(X>0) = \int_{0}^{1} f_X(x) dx = \frac{1}{2}$

$P(−0.5<X<0.5) = \int_{-0.5}^{0.5} f_X(x) dx = \frac{11}{16} = 0.6875$

$P(|X|>0.25) = 1 - \int_{-0.25}^{0.25} f_X(x) dx = \frac{81}{128} = 0.6328125$

### Pregunta C

Hallar $F_X$, la función de distribución acumulada de $X$.

$F_X = \int f_X(x) dx = \int 0.75(1-x^2) dx = 0.75 \int 1 - x^2 dx = 0.75 (x - \frac{x^3}{3})$

### Pregunta D

Graficar $f_X$ y $F_X$.

```r
f = function(x) 0.75 * (1 - x^2)
F = function(x) 0.75 * (x - x^3/3)

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=f) +
    xlim(-1, 1) +
    ylab('f')

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=F) +
    xlim(-1, 1) +
    ylab('F')
```

## Ejercicio 2

Sea $X$ una v.a. continua con función de distribución

$
F_X(x) =
\begin{cases}
  0 &\text{si } x < 0 \\
  \frac{x^3}{\theta} &\text{si } 0 \leq x < 2 \\
  1 &\text{si } 2 \leq x
\end{cases}
$

### Pregunta A

¿Cuál es el valor de $\theta$?

$\lim_{x \to 2^-} F_X(x) = 1 \iff \lim_{x \to 2^-} \frac{x^3}{\theta} = 1 \iff \frac{2^3}{\theta} = 1 \iff \theta = 8$

### Pregunta B

Calcular usando $F_X(x)$:

$P(X \leq 1) = F_X(1) = \frac{1}{8}$

$P(0.5 \leq X \leq 1) = F_X(1) - F_X(0.5) = \frac{1}{8} - \frac{1}{64} = \frac{7}{64}$

$P(0.5 \leq X \leq 1 | X < 1) = \frac{\frac{7}{64}}{\frac{1}{8}} = \frac{7}{8}$

### Pregunta C

Hallar la mediana $\tilde\mu$ de esta distribución.

$F_X(\tilde\mu) = \frac{1}{2} \iff \frac{\tilde\mu^3}{8} = \frac{1}{2} \iff \tilde\mu^3 = 4 \iff \tilde\mu = \sqrt[3]{4}$

### Pregunta D

Encontrar la función de densidad $f_X(x)$.

$f_X(x) = \frac{\partial}{\partial x} F_X(x) = \frac{3x^2}{8}$

### Pregunta E

Graficar $f_X$ y $F_X$. Agregar la recta vertical $x = \tilde\mu$.

```r
f = function(x) 3 * x^2 / 8
F = function(x) x^3 / 8
u = 4^(1/3)

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=f) +
    xlim(0, 2) +
    geom_vline(xintercept=u, linetype='dashed', color='blue') +
    ylab('f')

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=F) +
    xlim(0, 2) +
    geom_vline(xintercept=u, linetype='dashed', color='blue') +
    ylab('F')
```

## Ejercicio 3

Sea $U$ una v.a. con distribución $\mathcal{U}[0, \theta]$, para $\theta > 0$.

### Pregunta A

Hallar la función de distribución acumulada de $U$.

$F_U(u) = \frac{u - 0}{\theta - 0} = \frac{u}{\theta} I_{[0, \theta]}(u)$

### Pregunta B

Si $P(1 \leq U \leq 3) = 0.5$, ¿qué valores puede tomar $\theta$?

Caso $\theta \geq 3$

$P(1 \leq U \leq 3) = 0.5 \iff F_U(3) - F_U(1) = 0.5 \iff \frac{3}{\theta} - \frac{1}{\theta} = 0.5 \iff \frac{2}{\theta} = 0.5 \iff \theta = 4$

Caso $\theta \leq 3$

$P(1 \leq U \leq 3) = 0.5 \iff F_U(3) - F_U(1) = 0.5 \iff 1 - \frac{1}{\theta} = 0.5 \iff \frac{1}{\theta} = 0.5 \iff \theta = 2$

Caso $\theta \leq 1$

$P(1 \leq U \leq 3) = 0.5 \iff F_U(3) - F_U(1) = 0.5 \iff 1 - 1 = 0 \neq 0.5$

### Pregunta C

Para $\theta$ hallados en el item b) estimar la $P(U^2 < 2)$.

$P(U^2 < u) = P(U < \sqrt{u}) = F_U(\sqrt{u}) = \frac{\sqrt{u}}{\theta}$

Caso $\theta = 4$

$P(U^2 < 2) = \frac{\sqrt{2}}{4}$

Caso $\theta = 2$

$P(U^2 < 2) = \frac{\sqrt{2}}{2}$

## Ejercicio 4

Consideremos una v.a. $Y$ con función de densidad

$
f_Y(y) =
\begin{cases}
    \frac{1}{25}y & 0 \leq y < 5 \\
    \frac{2}{5} - \frac{1}{25}y & 5 \leq y < 10 \\
    0 & \text{en otro caso}
\end{cases}
$

### Pregunta A

Calcular la función de distribución de $Y$.

Caso $0 \leq y < 5$

$F_Y(y) = \int_{0}^{y} f_Y(t) dt = \int_{0}^{y} \frac{1}{25}t dt = \frac{1}{25} \Big[ \frac{t^2}{2} \Big]_0^y = \frac{1}{25} \frac{y^2}{2} = \frac{y^2}{50}$

Caso $5 \leq y < 10$

$
\begin{align*}
F_Y(y)
&= \int_{0}^{5} f_Y(t) dt + \int_{5}^{y} f_Y(t) dt \\
&= \tfrac{5^2}{50} + \int_{5}^{y} \tfrac{2}{5} - \tfrac{1}{25}t dt \\
&= \tfrac{1}{2} + \Big[ \tfrac{2}{5}t - \tfrac{t^2}{50} \Big]_5^y \\
&= \tfrac{1}{2} + \left( \tfrac{2}{5}y - \tfrac{y^2}{50} \right) - \left( \tfrac{2}{5}5 - \tfrac{5^2}{50} \right) \\
&= \tfrac{1}{2} + \tfrac{20y - y^2}{50} - \tfrac{3}{2} \\
&= \tfrac{20y - y^2}{50} - 1
\end{align*}
$

$
F_Y(y) =
\begin{cases}
    \tfrac{y^2}{50} & 0 \leq y < 5 \\
    \tfrac{20y - y^2}{50} - 1 & 5 \leq y < 10 \\
    0 & \text{en otro caso}
\end{cases}
$

### Pregunta B

Calcular $E(Y)$ y $V(Y)$.

$
\begin{align*}
E(Y)
&= \int_{-\infty}^{\infty} y f_Y(y) dy \\
&= \int_0^5 \tfrac{1}{25}y^2 dy + \int_5^{10} \tfrac{2}{5}y - \tfrac{1}{25}y^2 dy \\
&= \Big[ \tfrac{1}{75}y^3 \Big]_0^5 + \Big[ \tfrac{1}{5}y^2 - \tfrac{1}{75}y^3 \Big]_5^{10} \\
&= \tfrac{1}{75}5^3 + \left( \tfrac{1}{5}10^2 - \tfrac{1}{75}10^3 \right) - \left( \tfrac{1}{5}5^2 - \tfrac{1}{75}5^3 \right) \\
&= 5
\end{align*}
$

$
\begin{align*}
V(Y)
&= E(Y^2) - (E(Y))^2 \\
&= E(Y^2) - \mu_Y^2 \\
&= \int_{-\infty}^{\infty} y^2 f_Y(y) dy - \mu_Y^2 \\
&= \int_0^5 \tfrac{1}{25}y^3 dy + \int_5^{10} \tfrac{2}{5}y^2 - \tfrac{1}{25}y^3 dy - \mu_Y^2 \\
&= \Big[ \tfrac{1}{100}y^4 \Big]_0^5 + \Big[ \tfrac{2}{15}y^3 - \tfrac{1}{100}y^4 \Big]_5^{10} - \mu_Y^2 \\
&= \tfrac{1}{100}5^4 + \left( \tfrac{2}{15}10^3 - \tfrac{1}{100}10^4 \right) - \left( \tfrac{2}{15}5^3 - \tfrac{1}{100}5^4 \right) - \mu_Y^2 \\
&= \tfrac{175}{6} - 5^2 \\
&= \tfrac{25}{6}
\end{align*}
$

### Pregunta C

Calcular $E(\frac{1}{Y})$. ¿Qué conclusión saca respecto a la relación entre $E(\frac{1}{Y})$ y $\frac{1}{E(Y)}$.

$
\begin{align*}
E(\tfrac{1}{Y})

&= \Big[ \tfrac{1}{25}y \Big]_0^5 + \Big[ \tfrac{2}{5}ln(y) - \tfrac{1}{25}y \Big]_5^{10} \\

&= \tfrac{2}{5}ln(10) - \tfrac{2}{5}ln(5) \\
&= \tfrac{2}{5}ln(\tfrac{10}{5}) \\
&= \tfrac{2}{5}ln(2) \\
&\approx 0.27725
\end{align*}
$

## Ejercicio 5

Se eligen $n$ puntos al azar en el intervalo $[0, 1]$ de forma independiente (con distribución uniforme). Sea $X$ = Cantidad de puntos que caen en el intervalo $[0, p]$ $(0 < p < 1)$. ¿Qué distribución tiene $X$?

Sea $Y$ una v.a. continua con distribución $\mathcal{U}(0, 1)$.

$P(Y < p) = F_Y(p) = p$

Es decir, la probabilidad de que un punto elegido caiga en el intervalo $[0, p]$ es $p$.

Por cada punto elegido, hay solo dos posibilidades: o cae en el intervalo $[0, p]$ o cae en su complemento $[p, 1-p]$.

Por lo tanto: $X \sim Bi(n, p)$

## Ejercicio 7

Sea $X$ una v.a. con distribución $N(5, 0.25)$.

$X \sim N(5, 0.25) \implies Z = \frac{X - 5}{\sqrt{0.25}} = \frac{X - 5}{0.5} \sim N(0, 1)$

### Pregunta A

$P(4.75 \leq X \leq 5.5) = F_X(5.5) - F_X(4.75) = \Phi(\frac{5.5 - 5}{0.5}) - \Phi(\frac{4.75 - 5}{0.5}) = \Phi(1) - \Phi(-0.5)$

```r
# Usando la versión estandarizada de la v.a. X (lo que habría que hacer si se quiere calcular esto por tabla)
pnorm(1, 0, 1) - pnorm(-0.5, 0, 1)

# Usando directamente la v.a. X original, ya que R permite especificar los parámetros μ y σ
mu = 5
sigma = sqrt(0.25)
pnorm(5.5, mu, sigma) - pnorm(4.75, mu, sigma)
```

### Pregunta B

$P(|X| > 5.25) = 1 - P(|X| \leq 5.25) = 1 - P(-5.25 \leq X \leq 5.25) = 1 - (\Phi(\frac{5.25 - 5}{0.5}) - \Phi(\frac{-5.25 - 5}{0.5})) = 1 - (\Phi(0.5) - \Phi(-20.5))$

```r
# Forma 1
1 - (pnorm(0.5, 0, 1) - pnorm(-20.5, 0, 1))

# Forma 2
mu = 5
sigma = sqrt(0.25)
1 - (pnorm(5.25, mu, sigma) - pnorm(-5.25, mu, sigma))
```

### Pregunta C

$c$ tal que $P(|X−5| \leq c) = 0.9$

$
\begin{align*}
P(|X−5| \leq c)
&= P(-c \leq X−5 \leq c) \\
&= P(-c+5 \leq X \leq c+5) \\
&= F_X(c+5) - F_X(-c+5) \\
&= \Phi(\tfrac{c+5-5}{0.5}) - \Phi(\tfrac{-c+5-5}{0.5}) \\
&= \Phi(\tfrac{c}{0.5}) - \Phi(\tfrac{-c}{0.5}) \\
\end{align*}
$

Por simetría de la distribución normal standard: $\Phi(\tfrac{c}{0.5}) - \Phi(\tfrac{-c}{0.5}) = 0.9 \iff \Phi(\tfrac{c}{0.5}) - \Phi(0) = 0.45 \iff \Phi(\tfrac{c}{0.5}) = 0.45 + \Phi(0) \iff \Phi(\tfrac{c}{0.5}) = 0.95$

Luego: $\Phi(\tfrac{c}{0.5}) = 0.95 \iff \Phi^{-1}(0.95) * 0.5 = c$

```r
qnorm(0.95, 0, 1) * 0.5
```

### Pregunta D

El 90-percentil de $X$.

$P(X \leq x_{0.9}) = F_X(x_{0.9}) = 0.9 \iff \Phi(\tfrac{x_{0.9}-5}{0.5}) = 0.9 \iff \Phi(0.9)^{-1} = \tfrac{x_{0.9}-5}{0.5} \iff \Phi(0.9)^{-1} * 0.5 + 5 = x_{0.9}$

```r
qnorm(0.9, 0, 1) * 0.5 + 5
```
