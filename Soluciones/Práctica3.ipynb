## Ejercicio 1

Sea $X$ una v.a. con función de densidad:

$
f_X(x) =
\begin{cases}
  0.75(1-x^2) & -1 \leq x \leq 1 \\
  0 & \text{en otro caso}
\end{cases}
$

### Pregunta A

Verificar que $f_X$ es realmente una función de densidad.

Una función es una función de densidad (PDF) si cumple con 2 cosas:

1. $f_X(x) \geq 0$ para todo $-1 \leq x \leq 1$
2. $\int_{-1}^{1} f_X(x) dx = 1$

$f_X(x) = 0.75(1-x^2)$

$0.75(1-x^2) \geq 0 \iff 1-x^2 \geq 0 \iff 1 \geq x^2 \iff 1 \geq \pm x \iff -1 \leq x \leq 1$

$\int_{-1}^{1} f_X(x) dx = \int_{-1}^{1} 0.75(1-x^2) dx = 1$

### Pregunta B

Calcular: 

$P(X>0) = \int_{0}^{1} f_X(x) dx = \frac{1}{2}$

$P(−0.5<X<0.5) = \int_{-0.5}^{0.5} f_X(x) dx = \frac{11}{16} = 0.6875$

$P(|X|>0.25) = 1 - \int_{-0.25}^{0.25} f_X(x) dx = \frac{81}{128} = 0.6328125$

### Pregunta C

Hallar $F_X$, la función de distribución acumulada de $X$.

$F_X = \int f_X(x) dx = \int 0.75(1-x^2) dx = 0.75 \int 1 - x^2 dx = 0.75 (x - \frac{x^3}{3})$

### Pregunta D

Graficar $f_X$ y $F_X$.

```r
f = function(x) 0.75 * (1 - x^2)
F = function(x) 0.75 * (x - x^3/3)

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=f) +
    xlim(-1, 1) +
    ylab('f')

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=F) +
    xlim(-1, 1) +
    ylab('F')
```

$
F_X(x) =

### Pregunta A

¿Cuál es el valor de $\theta$?

$\lim_{x \to 2^-} F_X(x) = 1 \iff \lim_{x \to 2^-} \frac{x^3}{\theta} = 1 \iff \frac{2^3}{\theta} = 1 \iff \theta = 8$

### Pregunta B

Calcular usando $F_X(x)$:

$P(X \leq 1) = F_X(1) = \frac{1}{8}$

$P(0.5 \leq X \leq 1) = F_X(1) - F_X(0.5) = \frac{1}{8} - \frac{1}{64} = \frac{7}{64}$

$P(0.5 \leq X \leq 1 | X < 1) = \frac{\frac{7}{64}}{\frac{1}{8}} = \frac{7}{8}$

### Pregunta C

Hallar la mediana $\tilde\mu$ de esta distribución.

$F_X(\tilde\mu) = \frac{1}{2} \iff \frac{\tilde\mu^3}{8} = \frac{1}{2} \iff \tilde\mu^3 = 4 \iff \tilde\mu = \sqrt[3]{4}$

### Pregunta D

Encontrar la función de densidad $f_X(x)$.

$f_X(x) = \frac{\partial}{\partial x} F_X(x) = \frac{3x^2}{8}$

### Pregunta E

Graficar $f_X$ y $F_X$. Agregar la recta vertical $x = \tilde\mu$.

```r
f = function(x) 3 * x^2 / 8
F = function(x) x^3 / 8
u = 4^(1/3)

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=f) +
    xlim(0, 2) +
    geom_vline(xintercept=u, linetype='dashed', color='blue') +
    ylab('f')

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=F) +
    xlim(0, 2) +
    geom_vline(xintercept=u, linetype='dashed', color='blue') +
    ylab('F')
```

Se eligen $n$ puntos al azar en el intervalo $[0, 1]$ de forma independiente (con distribución uniforme). Sea $X$ = Cantidad de puntos que caen en el intervalo $[0, p]$ $(0 < p < 1)$. ¿Qué distribución tiene $X$?

Sea $Y$ una v.a. continua con distribución $\mathcal{U}(0, 1)$.

$P(Y < p) = F_Y(p) = p$

Es decir, la probabilidad de que un punto elegido caiga en el intervalo $[0, p]$ es $p$.

\begin{align*}

### Pregunta A

## Ejercicio 15

El tiempo de caída de un sistema se define como la fracción de tiempo que el sistema no está operativo debido a una falla del hardware o del software. Supongamos que $T$ = tiempo de caída de un sistema en horas es una variable aleatoria con función de densidad dada por:

$f_T(t) = te^{-\tfrac{t^2}{2}} I_{[0, \infty]}(t)$

### Pregunta A

Deduzca la función de distribución acumulada de $T$.

$F_T(x) = \int_0^x te^{-\tfrac{t^2}{2}} dx = \Big[ -e^{-\tfrac{t^2}{2}} \Big]_0^x = 1 - e^{-\tfrac{x^2}{2}}$

```r
f = function(t) t * exp(-t^2 / 2)
F = function(t) 1 - exp(-t^2 / 2)

ggplot(data = data.frame(x=0), mapping=aes(x=x)) +
    stat_function(fun=f, color='red') +
    stat_function(fun=F, color='blue') +
    xlim(0, 4)
```

$P(T < 1) = F_T(1) = 1 - e^{-\tfrac{1^2}{2}} \approx 0.393469$

### Pregunta C

Supongamos que al caer el sistema, el tiempo que tarda un usuario en recuperar su trabajo es una función creciente del tiempo de caída, digamos $T^2$. Deduzca la función de densidad de $T^2$.

$F_{T^2}(t) = P(T^2 \leq t) = P(T \leq \sqrt{t}) = F_T(\sqrt{t})$

## Ejercicio 16

La vida útil (en meses) de un componente electrónico es una v.a. $V$ con distribución $\mathcal{E}(\lambda)$, tal que $P(V > 20) = 0.449$.

### Pregunta A

Hallar $E(V)$ y $Var(V)$.

$P(V > 20) = 1 - P(V \leq 20) = 1 - F_V(20) = 1 - (1 - e^{-20\lambda}) = e^{-20\lambda} = 0.449 \iff \lambda = -\tfrac{ln(0.449)}{20} \iff \lambda \approx 0.04$

$E(V) = \tfrac{1}{\lambda} = \tfrac{1}{0.04} \approx 25$

$Var(V) = \tfrac{1}{\lambda^2} = \tfrac{1}{0.04^2} \approx 625$

### Pregunta B

Hallar la probabilidad de que la vida útil de uno de estos componentes sea mayor que 10 meses.

$P(V > 10) = 1 - P(V \leq 10) = 1 - F_V(10) \approx 0.67032$

### Pregunta C

Si se sabe que uno de estos componentes dura más de 20 meses, ¿cuál es la probabilidad de que dure más de 30 meses? Comparar con (b).

Por la propiedad de falta de memoria de la distribución exponencial:

$P(V > 30 | V > 20) = P(V > 10) \approx 0.67032$

### Pregunta D

$P(V > t + s | V > t) = P(V > s)$

## Ejercicio 18

Un sistema consta de 5 componentes electrónicos como los del Ejercicio 16, conectados en serie. Al fallar uno cualquiera de éstos, automáticamente se desconecta el sistema. Se supone que los tiempos de vida de los componentes son independientes. Es decir, si se definen los eventos:

$A_i$ = {el i-ésimo componente dura por lo menos hasta el instante $t$}

con $i = 1, \dots , 5$, estos eventos son independientes.

Sea $X$ el momento en el cual el sistema se desconecta.

### Pregunta A

Escribir el evento $\{X \geq t\}$ en función de los $A_i$.

$\{X \geq t\} = \bigcap_{i=1}^5 A_i$

Es una intersección porque los $A_i$ indican que el componente i-ésimo duró hasta por los menos el instante $t$. Entonces si $X$ representa el momento en el cual el sistema se desconecta, necesitamos que todos los componentes duren hasta por lo menos el instante $t$. A partir de ahí, cuando al menos 1 componente falla ya se desconecta el sistema.

### Pregunta B

Usar la independencia de los $A_i$ para calcular $P(X \geq t)$.

$V$ = vida útil del componente electrónico

$V \sim \mathcal{E}(0.04)$

$P(A_i) = P(V \geq t) = e^{-0.04t}$

$P(X \geq t) = P(\bigcap_{i=1}^5 A_i) = \prod_{i=1}^5 P(A_i) = \prod_{i=1}^5 e^{-0.04t} = (e^{-0.04t})^5 = e^{-0.2t}$

### Pregunta C

Hallar las funciones de distribución y de densidad de $X$. ¿A qué familia pertenece esta distribución?

$P(X \geq t) = e^{-0.2t} \iff P(X \leq t) = 1 - e^{-0.2t} \iff F_X(t) = 1 - e^{-0.2t}$

$\tfrac{\partial}{\partial x} F_X(t) = f_X(t) = 0.2e^{-0.2t}$

$\implies X \sim \mathcal{E}(0.2)$

$X$ = tiempo de compilación en segundos de un bloque

### Pregunta A

Calcule la probabilidad de que alguno de los tres bloques tenga un tiempo de compilación superior a los 2 segundos.

$Y$ = cantidad de bloques con tiempo de compilación superior a 2 segundos

$Y \sim Bi(3, P(X \geq 2))$

$P(Y \geq 1)$

```r
pbinom(0, 3, pexp(2, 1, lower.tail=FALSE), lower.tail=FALSE)
```
